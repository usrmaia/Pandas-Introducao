```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
aracaju = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Aracaju.xlsx")
fortaleza = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Fortaleza.xlsx")
natal = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Natal.xlsx")
recife = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Recife.xlsx")
salvador = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Salvador.xlsx")
df = pd.concat([aracaju, fortaleza, natal, recife, salvador])
df["Receitas"] = df["Vendas"].mul(df["Qtde"])
df["Ano"] = df["Data"].dt.year
df["Mês"] = df["Data"].dt.month
df["Dia"] = df["Data"].dt.day
```

```python
# Basicamente conta as ocorrencias
df["LojaID"].value_counts()
```

```python
df["LojaID"].value_counts().plot.bar()
```

```python
df["LojaID"].value_counts(ascending=True).plot.barh();
```

```python
receita_por_ano = df.groupby("Ano")["Receitas"].sum()
receita_por_ano.plot.pie()
```

```python
df["Cidade"].value_counts()
```

```python
df["Cidade"].value_counts().plot.bar(title="Total de Vendas por Cidade", color="red")
plt.xlabel("Cidades")
plt.ylabel("Vendas")
```

## Alterando estilo do gráfico

```python
plt.style.use("ggplot") # Muda o background
df.groupby("Mês")["Qtde"].sum().plot(title="Quantidade de Vendas por Mês", marker="o") # Estilo default de plot são linhas
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.legend()
```

```python
plt.hist(df["Qtde"], color="blue")
```

```python
df_2019 = df[df["Ano"] == 2019]
plt.scatter(x=df_2019["Dia"], y=df_2019["Receitas"])
```

## Salvando os gráficos produzidos

```python
plt.scatter(x=df_2019["Dia"], y=df_2019["Receitas"])
plt.savefig("Gráfico de dispersão: Receita por ano")
```
