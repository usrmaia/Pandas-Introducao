```python
import pandas as pd
```

```python
aracaju = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Introdução ao Pandas/datasets/Aracaju.xlsx")
fortaleza = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Introdução ao Pandas/datasets/Fortaleza.xlsx")
natal = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Introdução ao Pandas/datasets/Natal.xlsx")
recife = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Introdução ao Pandas/datasets/Recife.xlsx")
salvador = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Introdução ao Pandas/datasets/Salvador.xlsx")
```

### As vezes, quando não vamos realizar operações com determinada coluna de uma tabela, pode ser interessante deixa-la com tipo object (equivalente a string).

## Identificando e tratando celulas nulas

```python
 df.isnull().sum()
```

Suponda que há valores nulos em vendas, podemos tratalos de algumas formas:

```python
m = df["Vendas"].mean()
df["Vendas"].fillna(m, inplace=True)
# df["Vendas"].fillna(0, inplace=True)
# df.dropna(inplace=True) # Dropar linhas com valores nulos
# df.dropna(subset=["Vendas"], inplace=True) # Apenas com base na coluna vendas
# df.dropna(how="all", inplace=True) # Dropa caso todas as colunas sejam nulas
```

```python
 df.isnull().sum()
```

## Criando novas colunas

```python
df["Receitas"] = df["Vendas"].mul(df["Qtde"])
```
