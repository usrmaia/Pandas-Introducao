```python
import pandas as pd
```

```python
aracaju = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Aracaju.xlsx")
fortaleza = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Fortaleza.xlsx")
natal = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Natal.xlsx")
recife = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Recife.xlsx")
salvador = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DIO - Introdução ao Pandas/datasets/Salvador.xlsx")
df = pd.concat([aracaju, fortaleza, natal, recife, salvador])
df["Receitas"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

## Ao se manipular datas em um df, nem sempre o tipo do objeto será Date, pode acontecer de ser int, float ou double

```python
df["Data"] = df["Data"].astype("int64")
```

## Transformando em date

```python
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
df.groupby(df["Data"].dt.year)["Receitas"].sum()
```

```python
df["Ano"] = df["Data"].dt.year
df["Mês"] = df["Data"].dt.month
df["Dia"] = df["Data"].dt.day
```

## Operações com data

```python
df["Data"].min()
```

```python
df["Diferênça de Dias"] = df["Data"] - df["Data"].min()
```

```python
# Criando Trimestre
df["Trimestre"] = df["Data"].dt.quarter
```

```python
# Filtrando as vendas
df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
